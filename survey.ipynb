```python
filter =(data['Country']=="India") &(data['RemoteWork']=="Remote") & (data['ConvertedCompYearly']>100000)
data[filter]
```

```python
#To know from where they have learnt
data[filter].LearnCode.value_counts()
```

```python
#To know from where they have learnt -Online
data[filter].LearnCodeOnline.value_counts()
```
